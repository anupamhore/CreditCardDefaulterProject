```python
def outlier_imputation(df,cols,types):
    try:
        if(len(cols) == len(types)):
            count = 0
            for col in cols:
                typestr = types[count]
                if typestr == "Gaussian":
                    upper_bound = df[col].mean() + 3 * df[col].std()
                    lower_bound = df[col].mean() - 3 * df[col].std()
                    df[col].clip(upper=upper_bound,inplace=True)
                    df[col].clip(lower=lower_bound,inplace=True)
                elif typestr == "Skewed":
                    IQR = df[col].quantile(0.75) - df[col].quantile(0.25)
                    upper_bound = df[col].quantile(0.75) + 1.5 * IQR
                    lower_bound = df[col].quantile(0.25) - 1.5 * IQR
                    df[col].clip(upper=upper_bound,inplace=True)
                    df[col].clip(lower=lower_bound,inplace=True)
                else:
                    IQR = df[col].quantile(0.75) - df[col].quantile(0.25)
                    upper_bound = df[col].quantile(0.75) + 3 * IQR
                    lower_bound = df[col].quantile(0.25) - 3 * IQR
                    df[col].clip(upper=upper_bound,inplace=True)
                    df[col].clip(lower=lower_bound,inplace=True)
                count  = count +1
            return df
        else:
            return 'Length of type and columns mismatch'
    except Exception as e:
        return Exception(e)
```

```python
cols_needing_outlierTreatment = ['AGE','BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6',
                                             'EDUCATION','LIMIT_BAL','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6',
                                             'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']
distributionTypes =['Skewed','Highly Skewed','Highly Skewed','Highly Skewed','Highly Skewed','Highly Skewed','Highly Skewed',
                                'Skewed','Skewed','Gaussian','Gaussian','Gaussian','Gaussian','Gaussian','Gaussian',
                                'Highly Skewed','Highly Skewed','Highly Skewed','Highly Skewed','Highly Skewed','Highly Skewed']

df = outlier_imputation(df,cols_needing_outlierTreatment,distributionTypes)
```

### Find the relationship between independant variables and dependant variables

- Check which variables have good explainatory powers

```python
import scipy.stats as stats
```

```python
#Find chi-square Test
chisq_df = pd.DataFrame()
for feature in categorical_features:
    cross_tab = pd.crosstab(df[feature],df['default.payment.next.month'],margins=False)
    tmp = pd.DataFrame([feature,stats.chi2_contingency(observed=cross_tab)[0],stats.chi2_contingency(observed=cross_tab)[1]]).T
    tmp.columns=['Feature','Chi-Square','p-Value']
    chisq_df = pd.concat([chisq_df,tmp],axis=0,ignore_index=True)
    
chisq_df
```

```python
tstats_df = pd.DataFrame()
numerical_features1 = [i for i in numerical_features if i not in ['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']]
for feature in numerical_features1:
    tstats = stats.ttest_ind(df[df['default.payment.next.month'] == 1][feature],df[df['default.payment.next.month'] == 0][feature])
    temp = pd.DataFrame([feature,tstats[0],tstats[1]]).T
    temp.columns=['Feature','T-Statistics','p-Value']
    tstats_df = pd.concat([tstats_df,temp],axis=0,ignore_index=True)

print(tstats_df)
```

#### Factor Plot for categorical variable compared to Target Variable

```python
def plotpercentageGraph(col,df):
    count = df[[col,'default.payment.next.month']].groupby([col,'default.payment.next.month']).size()
    percentage = count.groupby(level=[0]).apply(lambda x: x/x.sum()).reset_index()
    percentage.columns = [col,'Defaulter','Percentage']
    return percentage
```

```python
for feature in categorical_features:
    percentage = plotpercentageGraph(feature,df)
    fig,axes = plt.subplots(1,1,figsize=(10,4))
    sns.catplot(x=feature,y=percentage['Percentage'],hue='Defaulter',data=percentage,size=4,kind='bar')
    plt.xlabel('Plot for ' + str(feature))
    plt.title('Factor plor for ' + str(feature) + " Defaulter")
    plt.close(1)
    plt.show()
```

#### Initial Analysis

- Percentage of defaulters is more for customers who are married or whose marriage identity is not known(may be divorced or single parent)

- Though the count of female credit owners are more but we can see males to be slightly more defaulters than females

- For April, May, July, August and September month, we can observe that most of the customers who have payment delay for more than 2 months have more than 50% chance of being defaulters.

- For the month of June, we can observe the customers with payment delay for more than or equal to 1 month, have more than 50% chance of being defaulters.

### Check if any bucketing is required for any variable ( Continous/Discrete to categorical)

```python
for feature in df.columns.difference(['default.payment.next.month']):
    binned = pd.cut(df[feature], bins=10, labels= list(range(1,11)))
    binned = binned.dropna()
    ser = df.groupby(binned)['default.payment.next.month'].sum() / (df.groupby(binned)['default.payment.next.month'].count() - df.groupby(binned)['default.payment.next.month'].sum())
    ser = np.log(ser)
    fig, axes = plt.subplots(figsize=(10,4))
    sns.barplot(x=ser.index, y=ser)
    plt.ylabel('Log Odds Ratio')
    plt.title(str('Logit Plot for identifying if the bucketing is required or not for the variable ') + str(feature))
    
```

#### Initial Analysis

- It is quiet clear from the above graph that all the variables with PAY_* can be binned and can be made as categorical
